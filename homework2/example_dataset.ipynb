```python
import numpy as np
```

```python
import matplotlib.pyplot as plt

plt.style.use('seaborn')
```

```python
n_points = 100
```

```python
sigma = .06
x = 0.5 * np.random.randn(n_points, 2)

y = np.random.randn(n_points, 2) 
y /= np.linalg.norm(y, axis=1, keepdims=True)
y = y + sigma * np.random.randn(n_points, 2)
y[:, 0] *= .3
y[:, 0] += 1.5
y[:, 1] *= 1.3

```

```python
plt.scatter(*x.T)
plt.scatter(*y.T)
plt.axis('scaled')
```

```python
z = np.vstack([x, y])
z.shape
```

```python
labels = np.zeros((z.shape[0]))

labels[n_points:] = 1
```

```python
labels
```

```python
np.savetxt("data.dat", np.hstack([z, labels[:, None]]))
```
