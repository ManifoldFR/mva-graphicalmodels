```python
import numpy as np
```

```python
import matplotlib.pyplot as plt

plt.style.use('seaborn')
```

```python
ns1 = 200
ns2 = 300
```

```python
sigma = .16
x = 0.6 * np.random.randn(ns1, 2)
x[:, 0] *= .4
x[:, 1] *= .5
theta = np.pi / 12
x[:,0] = np.cos(theta)*x[:,0] + np.sin(theta)*x[:,1]
x[:,1] = -np.sin(theta)*x[:,0] + np.cos(theta)*x[:,1]


y = np.random.randn(ns2, 2) 
y /= np.linalg.norm(y, axis=1, keepdims=True)
y = y + sigma * np.random.randn(ns2, 2)
y[:, 0] *= .18
y[:, 0] += .8
y[:, 1] *= 1.

```

```python
plt.scatter(*x.T, s=15)
plt.scatter(*y.T, s=15)
plt.axis('scaled')
```

```python
z = np.vstack([x, y])
z.shape
```

```python
labels = np.zeros((z.shape[0]))
labels[ns1:] = 1
```

```python
np.savetxt("data.dat", np.hstack([z, labels[:, None]]))
```
